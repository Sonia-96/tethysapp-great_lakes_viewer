```python
import pandas as pd
```

```python
output_dir = 'workspaces/app_workspace/data'
```

```python
df_data = pd.read_csv(f'{output_dir}/GLHYD_data_english.csv', skiprows=12)

def convert_time(cols):
    month, year = cols['month'], cols['year']
    month_to_number = {
        'jan': '01', 'feb': '02', 'mar': '03', 'apr': '04',
        'may': '05', 'jun': '06', 'jul': '07', 'aug': '08',
        'sep': '09', 'oct': '10', 'nov': '11', 'dec': '12'
    }
    return f'{year}-{month_to_number[month]}-01'
df_data['Month'] = df_data[['month', 'year']].apply(convert_time, axis=1)
df_data = df_data.drop(columns=['month', 'year'])
df_data = df_data.set_index('Month')
df_data
```

```python
for col in df_data.columns:
    df_data[col].to_csv(f'{output_dir}/Lake {col}.csv')
```
